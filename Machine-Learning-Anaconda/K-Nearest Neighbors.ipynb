```python
#Split the data into training and test set
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 0)
```

```python
#import and instantiate the class.
#Set number of neighbors to use
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors = 3)
```

```python
#firt classidier using the training set
clf.fit(X_train, y_train)
```

```python
print("Test set predictions: {}".format(clf.predict(X_test)))
```

```python
print("Test set accuracy: {:.2f}".format(clf.score(X_test, y_test)))
```

```python
import matplotlib.pyplot as plt


#visualization of the decision boundaries
fig, axes = plt.subplots(1,3, figsize=(10,3))

print("fig: {}".format(fig))
print("axes: {}".format(axes))

for n_neighbors, ax in zip([1,3,9], axes):
    #fit method return the object self, so it can be instantiated and fit in one line
    #.fit(X,y): fit the model using X as training data and y as target value
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X,y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:,0], X[:,1], y, ax=ax)
    ax.set_title("{} neighbor(s)".format(n_neighbors))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
    
axes[0].legend(loc=0)
```
